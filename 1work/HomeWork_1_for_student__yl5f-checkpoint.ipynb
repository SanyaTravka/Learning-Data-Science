Итак, прежде чем приступать к выполнению задания, познакомим Вас с функцией python **assert**

**assert** -- это Ваш будущий друг и товарищ, который будет сопровождать Вас по ходу всего курса. Эта функция нужна для того, чтобы проверить верно ли реализован код. 

Структура этой функции выглядит следующим образом:

<code>assert condition, message</code>

**assert** -- специальное слово python для вызова функции.

**condition** -- Условие, которые мы хотим проверить. В данном случае мы хотим проверить, что <code>sum(df["Д"]) == 2</code>, т.е. мы хотим убедиться, действительно ли фитнесом занимается всего 2 человека.

**message** -- Это просто строка, которая будет выводиться, если condition **НЕ** будет выполняться.



Функция **assert** будет встречаться у Вас по ходу всего кода и помогать Вам проверять верно ли реализован какой-то конкретный участок кода.

**Чтобы пройти эту проверку, просто запустите ячейку кода с assert ничего не меняя в ней.**

* Если **участок кода реализован правильно**, то ячейка выполнится без ошибок. После этого Вы можете продолжать выполнять задание, зная что предыдущие шаги у Вас реализованы правильно.

* Если **участок кода реализован НЕправильно**, то ячейка закончит выполнение с ошибкой <code>AssertionError: message</code>. Это сообщение говорит Вам, что Вы где-то совершили ошибку в реализации и сделали что-то неправильно. Тогда Вам нужно еще раз внимательно посмотреть на код, найти в нем ошибку, исправить её, а после этого снова пройти эту проверку. Как только вы её пройдете, Вы сможете приступать к выполнению следующих шагов.

```python
assert sum(df["Д"]) == 2, "Неправильно заполнена таблица"
```
