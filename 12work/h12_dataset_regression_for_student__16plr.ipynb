```python
import pandas as pd
```

```python
df = pd.read_csv("hw12_regression_dataset.csv", index_col=0)
```

```python
df
```

Данные считаются неправильными если:

* Имеются пропущенные значения NaN

* Возраст человека больше либо равен 100 г.

* Рост человека больше 200 см.

```python
df.info()
```

```python
df_incorrect1 = df[df["Пол"].isna()]
```

```python
df_incorrect1
```

```python
df_correct = df.dropna()
```

```python
df_correct.describe()
```

```python
df_incorrect2 = df_correct[df_correct["Возраст"]>=100]
```

```python
df_correct = df_correct[df_correct["Возраст"]<100]
```

```python
df_incorrect3 = df_correct[df_correct["Рост"]>200]
```

```python
df_correct = df_correct[df_correct["Рост"]<=200]
```

Для восстановления данных используйте среднее значение по столбцам "правильного датасета".

Для замены переменных в столбце "Пол" используйте следующую конструкцию: <code>map({"М" : 1, "Ж" : 0}) </code>

```python
df_correct["Пол"] = df_correct["Пол"].map({"М" : 1, "Ж" : 0})
```

```python
df_correct
```

```python
mean_gender = int(df_correct["Пол"].mean())
mean_age = int(df_correct["Возраст"].mean())
mean_height = int(df_correct["Рост"].mean())
(mean_gender,mean_age,mean_height)
```

```python
df_incorrect1 = df_incorrect1.fillna(mean_gender)
```

```python
df_incorrect1
```

```python
df_incorrect2["Возраст"] = mean_age
df_incorrect2["Пол"] = df_incorrect2["Пол"].map({"М" : 1, "Ж" : 0})
```

```python
df_incorrect2
```

```python
df_incorrect3["Рост"] = mean_height
df_incorrect3["Пол"] = df_incorrect3["Пол"].map({"М" : 1, "Ж" : 0})
```

```python
df_incorrect3
```

Команда <code>concat</code> позволяет объединить несколько таблиц в одну.

```python
df_final = pd.concat([df_correct,df_incorrect1,df_incorrect2,df_incorrect3])
```

```python
assert len(df_final) == len(df), "Ошибка! Размер финальной таблицы меньше исходнойю" 
```
